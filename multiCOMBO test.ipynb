```python
import multicombo
import deplacy
import graphviz

nlp=multicombo.load("en")

with open('/content/drive/MyDrive/Colab Notebooks/data/original.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()

for i, line in enumerate(lines):
    doc=nlp(line.strip())
    with open('/content/drive/MyDrive/Colab Notebooks/data/multicombo output/{}_conllu.txt'.format(i), 'w', encoding='utf-8') as f:
        f.write(multicombo.to_conllu(doc))
    graph_src = graphviz.Source(deplacy.dot(doc))
    graph_src.render(filename='{}_graph'.format(i), directory='/content/drive/MyDrive/Colab Notebooks/data/multicombo output/', cleanup=True)
```
