```python
%cd /content/drive/MyDrive/Tools/muss/

import os


def simplify_docs(inpath="data/original", outpath="data/simplified"):
    """Create files with simplified sentences
        
    inpath -- input dir; input files must have each sentence on a separate line.
    outpath -- output dir
    """

    files = os.listdir(inpath)
    for name in files:
        print(name)
        os.system(f"python scripts/simplify.py {inpath}/{name} {outpath}/{name}")


simplify_docs()
```
